- у циклі створити мінімум 3 об'єкти класу з різними вхідними даними

```python
class FactorialCalculator:
    def __init__(self, n):
        if n <= 0:
            raise ValueError("Число повинно бути додатним")
        self.n = n
        self._factorial_value = None
    
    def calculate(self):
        result = 1
        for i in range(1, self.n + 1):
            result *= i
        self._factorial_value = result
        return result
    
    @property
    def factorial(self):
        if self._factorial_value is None:
            self.calculate()
        return self._factorial_value
    
    def __str__(self):
        return f"Ім'я: Анна Рожанська\nГрупа: КН-32"


# Створення об'єктів класу у циклі
numbers = [3, 5, 7]
for num in numbers:
    calc = FactorialCalculator(num)
    print(f"Факторіал {num}! = {calc.factorial}")
    print(calc)
    print("-" * 30)
```
