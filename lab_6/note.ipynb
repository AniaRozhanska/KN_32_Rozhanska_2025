- –ø—Ä–æ—Ä–∞–º–∞ —è–∫–∞ –≤–∏–∑–Ω–∞—á–∞—î —á–∏ —Å—Ç—É–¥–µ–Ω—Ç –∑–∞–Ω—É–¥–±–≥—É–≤–∞–≤ –Ω–∞ –ª–µ–∫—Ü—ñ—ó

```python
class CheckSudentBored:
    def __init__(self, name, words_threshhold=50):
        self.__name = name          # –ü—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç
        self.__words_threshhold = words_threshhold  # –ü—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç
    
    def __generate_lecture_words(self):
        # –ü—Ä–∏–≤–∞—Ç–Ω–∏–π –º–µ—Ç–æ–¥ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü—ñ—ó –∫—ñ–ª—å–∫–æ—Å—Ç—ñ —Å–ª—ñ–≤
        import random
        return random.randint(0, 100)
    
    def __is_bored(self):
        lecture_words = self.__generate_lecture_words()
        return lecture_words > self.__words_threshhold
    
    def check_if_student_bored(self):
        if self.__is_bored():
            return f"–°—Ç—É–¥–µ–Ω—Ç {self.__name} –∑–∞–Ω—É–¥—å–≥—É–≤–∞–≤ –Ω–∞ –ª–µ–∫—Ü—ñ—ó."
        else:
            return f"–°—Ç—É–¥–µ–Ω—Ç {self.__name} –ù–ï –∑–∞–Ω—É–¥—å–≥—É–≤–∞–≤ –Ω–∞ –ª–µ–∫—Ü—ñ—ó."

b = CheckSudentBored("–í—ñ–∫—Ç–æ—Ä", words_threshhold=40)
b.check_if_student_bored()
```

# –ü—Ä–∏–∫–ª–∞–¥ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–Ω—è –≤ Python
- –Ω–∞–ø–∏—à–µ–º–æ –ø—Ä–æ–≥—Ä–∞–º—É —è–∫–∞ —ñ–º—ñ—Ç—É—î —Ä–æ–±–æ—Ç—É —Å—Ç–∞–Ω—Ü—ñ—ó –±–µ–ø–µ—Ä–µ–±—ñ–π–Ω–æ–≥–æ –∂–∏–≤–ª–µ–Ω–Ω—è
- –º–∏ —Ä–æ–∑—à–∏—Ä–∏–º–æ –¥–∞–Ω—É —Å—Ç—ñ–Ω—Ü—ñ—é —â–æ–± –º–æ–∂–Ω–∞ –±—É–ª–æ –≥–µ–Ω–µ—Ä—É–≤–∞—Ç–∏ –µ–ª–∫—Ç—Ä–∏–∫—É –≤—ñ–¥ —Å–æ–Ω—è—á–Ω–∏—Ö –ø–∞–Ω–µ–ª–µ–π, –≤—ñ—Ç—Ä—è–∫—ñ–≤ —Ç–∞ –¥–∏–∑–µ–ª—å–Ω–æ–≥–æ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞
- –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ –º–∏ –ø—Ä–∏–¥—É–º–∞—î–º–æ —Ñ—É—Ç—É—Ä–∏—Å—Ç–∏—á–Ω—ñ –¥–∂–µ—Ä–µ–ª–∞ –µ–Ω–µ—Ä–≥—ñ—ó

```python
class EnergyDevice:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.__max_capacity = capacity

    @property
    def info(self):
        return f"–ü—Ä–∏—Å—Ç—Ä—ñ–π: {self.name}, –Ñ–º–Ω—ñ—Å—Ç—å: {self.__max_capacity}"
    
    def recalculate_total_capacity(self, additional_capacity=0):
        self.__max_capacity += additional_capacity

    def charge_device(self):
        print(f"{self.name} –∑–∞—Ä—è–¥–∂–∞—î—Ç—å—Å—è –¥–æ —î–º–Ω–æ—Å—Ç—ñ {self.capacity}.")
        self.capacity += 10 if self.capacity < self.__max_capacity else self.__max_capacity
        return self.capacity
    
    def generate_energy(self):
        print(f"{self.name} –≥–µ–Ω–µ—Ä—É—î –µ–Ω–µ—Ä–≥—ñ—é –¥–æ —î–º–Ω–æ—Å—Ç—ñ {self.capacity}.")
        self.capacity -= 10 if self.capacity > 0 else 0
        return self.capacity

E = EnergyDevice("–ë–∞—Ç–∞—Ä–µ—è", 100)
E.generate_energy()
E.generate_energy()
E.generate_energy()
E.charge_device()
E.charge_device()
print(E.info)
```

```python
class SolarPanelDevice(EnergyDevice):
    def __init__(self, name, capacity, panel_area):
        super().__init__(name, capacity)
        self.panel_area = panel_area
        self.recalculate_total_capacity(self.__calculate_additional_capacity)
    
    @property
    def __calculate_additional_capacity(self):
        return self.panel_area * 2  # –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ –∫–æ–∂–µ–Ω –∫–≤–∞–¥—Ä–∞—Ç–Ω–∏–π –º–µ—Ç—Ä –¥–æ–¥–∞—î 2 –æ–¥–∏–Ω–∏—Ü—ñ —î–º–Ω–æ—Å—Ç—ñ
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, –ü–ª–æ—â–∞ –ø–∞–Ω–µ–ª—ñ: {self.panel_area} –º¬≤"

S = SolarPanelDevice("–°–æ–Ω—è—á–Ω–∞ –ü–∞–Ω–µ–ª—å", 100, 20)
print(S.info)
S.charge_device()
S.generate_energy()
S.generate_energy()
```

```python
class HybridSolarWindDevice(SolarPanelDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power):
        super().__init__(name, capacity, panel_area)
        self.wind_turbine_power = wind_turbine_power
        self.recalculate_total_capacity(self.__calculate_wind_additional_capacity)
        self.Test = "Hello"
    
    @property
    def __calculate_wind_additional_capacity(self):
        return self.wind_turbine_power * 3  # –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ –∫–æ–∂–µ–Ω –∫—ñ–ª–æ–≤–∞—Ç –¥–æ–¥–∞—î 3 –æ–¥–∏–Ω–∏—Ü—ñ —î–º–Ω–æ—Å—Ç—ñ
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, –ü–æ—Ç—É–∂–Ω—ñ—Å—Ç—å –≤—ñ—Ç—Ä—è–∫–∞: {self.wind_turbine_power} –∫–í—Ç"

H = HybridSolarWindDevice("–ì—ñ–±—Ä–∏–¥–Ω–∞ –°–æ–Ω—è—á–Ω–æ-–í—ñ—Ç—Ä—è–Ω–∞ –°–∏—Å—Ç–µ–º–∞", 150, 30, 5)
print(H.info)
H.generate_energy()
H.charge_device()
```

```python
class TeslaEtheriumDevice:
    def __init__(self, name, etherium_level):
        self.name = name
        self.capacity = etherium_level * 0.2 # –ü—Ä–∏–ø—É—Å—Ç–∏–º–æ, —â–æ –∫–æ–∂–µ–Ω —Ä—ñ–≤–µ–Ω—å –µ—Ñ—ñ—Ä—ñ—É–º—É –¥–æ–¥–∞—î 0.2 –æ–¥–∏–Ω–∏—Ü—ñ —î–º–Ω–æ—Å—Ç—ñ

class AdvancedTeslaDevice(HybridSolarWindDevice, TeslaEtheriumDevice):
    def __init__(self, name, capacity, panel_area, wind_turbine_power, etherium_level):
        HybridSolarWindDevice.__init__(self, name, capacity, panel_area, wind_turbine_power)
        TeslaEtheriumDevice.__init__(self, name, etherium_level)
        self.recalculate_total_capacity(self.__calculate_etherium_additional_capacity)
    
    @property
    def __calculate_etherium_additional_capacity(self):
        return self.capacity  # –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ —î–º–Ω—ñ—Å—Ç—å –∑ TeslaEtheriumDevice —è–∫ –¥–æ–¥–∞—Ç–∫–æ–≤—É —î–º–Ω—ñ—Å—Ç—å
    
    @property
    def info(self):
        base_info = super().info
        return f"{base_info}, –†—ñ–≤–µ–Ω—å –ï—Ñ—ñ—Ä—ñ—É–º—É: {self.capacity / 0.2}"

A = AdvancedTeslaDevice("–ü—Ä–æ—Å—É–Ω—É—Ç–∞ –¢–µ—Å–ª–∞ –°–∏—Å—Ç–µ–º–∞", 200, 40, 10, 50)
A.Hello = 10
print(A.Test, A.Hello)
```

# –ü—Ä–∏–∫–ª–∞–¥ –ê–±—Å—Ç—Ä–∞–∫—Ü—ñ—ó –≤ Python
- –ø–æ–¥–∏–≤–∏–º–æ—Å—å –∞–±—Å—Ç—Ä–∞—Ü—ñ—é –Ω–∞ –ø—Ä–∏–∫–ª–∞–¥—ñ LLM –º–æ–¥–µ–ª—ñ
- —Å–ø–æ—á–∞—Ç–∫—É –≤—Ç–æ—Ä–∏–º–æ –∞–±—Å—Ç—Ä–∞–∫—Ç–Ω–∏–π –∫–ª–∞—Å LLMModel –∑ –∞–±—Å—Ç—Ä–∞–∫—Ç–Ω–∏–º–∏ –º–µ—Ç–æ–¥–∞–º–∏ askLLm —Ç–∞ genereateResponse
- –¥–∞–ª—ñ —Å–ø—Ä–æ–±—É—î–º–æ —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –ø—Ä–æ—Ç–æ–∫–æ–ª —á–µ—Ä–µ–∑ typing.Protocol –¥–ª—è LLMModel

```python
from abc import ABC, abstractmethod

class LLMModel(ABC):
    @abstractmethod
    def askLLm(self, question):
        pass
    
    @abstractmethod
    def generateResponse(self, prompt):
        pass

class ItCollegeLLMModel(LLMModel):
    def __init__(self, model_name):
        self.model_name = model_name
    
    def askLLm(self, question):
        return f"{self.model_name} –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: {question}"
    
    def generateResponse(self, prompt):
        return f"{self.model_name} –≥–µ–Ω–µ—Ä—É—î –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –∑–∞–ø–∏—Ç: {prompt}"

class VoloymyrVLLMModel(LLMModel):
    def __init__(self, model_name):
        self.model_name = model_name
    
    def askLLm(self, question):
        return f"{self.model_name} –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: {question}"
    
    def generateResponse(self, prompt):
        return f"{self.model_name} –≥–µ–Ω–µ—Ä—É—î –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –∑–∞–ø–∏—Ç: {prompt}"
    

itllm = ItCollegeLLMModel("IT College LLM")
print(itllm.askLLm("–Ø–∫ –ø—Ä–∞—Ü—é—î LLM?"))
print(itllm.generateResponse("–Ø–∫—ñ —î –ø–µ—Ä–µ–≤–∞–≥–∏ LLM?"))

vvmodel = VoloymyrVLLMModel("Voloymyr V LLM")
print(vvmodel.askLLm("–©–æ —Ç–∞–∫–µ –ø–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º?"))
print(vvmodel.generateResponse("–Ø–∫ —Ä–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –ø–æ–ª—ñ–º–æ—Ä—Ñ—ñ–∑–º –≤ Python?"))
```

```python
from typing import Protocol
from abc import ABC, abstractmethod

class LLMModelProtocol(Protocol):
    @property
    def model_name(self) -> str:
        ...
    
    def tokenize(self, text: str) -> list:
        ...

class ModelInterface(ABC):
    @abstractmethod
    def askLLm(self, question):
        pass
    
    @abstractmethod
    def generateResponse(self, prompt):
        pass

class GPT3Model(LLMModelProtocol):
    def __init__(self, model_name):
        self._model_name = model_name
    
    @property
    def model_name(self) -> str:
        return self._model_name
    
    def tokenize(self, text: str) -> list:
        return text.split()  # –ü—Ä–æ—Å—Ç–∏–π –ø—Ä–∏–∫–ª–∞–¥ —Ç–æ–∫–µ–Ω—ñ–∑–∞—Ü—ñ—ó

class BERTModel(LLMModelProtocol):
    def __init__(self, model_name):
        self._model_name = model_name
    
    @property
    def model_name(self) -> str:
        return self._model_name
    
    def tokenize(self, text: str) -> list:
        return text.split()  # –ü—Ä–æ—Å—Ç–∏–π –ø—Ä–∏–∫–ª–∞–¥ —Ç–æ–∫–µ–Ω—ñ–∑–∞—Ü—ñ—ó

class ChatInterface(ModelInterface):
    def __init__(self, llm_model: LLMModelProtocol):
        self.llm_model = llm_model
    
    def askLLm(self, question):
        tokens = self.llm_model.tokenize(question)
        return f"{self.llm_model.model_name} –≤—ñ–¥–ø–æ–≤—ñ–¥–∞—î –Ω–∞ –ø–∏—Ç–∞–Ω–Ω—è: {tokens}"
    
    def generateResponse(self, prompt):
        tokens = self.llm_model.tokenize(prompt)
        return f"{self.llm_model.model_name} –≥–µ–Ω–µ—Ä—É—î –≤—ñ–¥–ø–æ–≤—ñ–¥—å –Ω–∞ –∑–∞–ø–∏—Ç: {tokens}"


for model in [GPT3Model("GPT-3"), BERTModel("BERT")]:
    chat_interface = ChatInterface(model)
    print(chat_interface.askLLm("–Ø–∫ –ø—Ä–∞—Ü—é—î LLM?"))
    print(chat_interface.generateResponse("–Ø–∫—ñ —î –ø–µ—Ä–µ–≤–∞–≥–∏ LLM?"))
```

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

circle = Circle(5)
print(circle.area())
```

# –í–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è –ø–∞—Ä–∞–¥–∏–≥–º –¥–ª—è —Å—Ç–≤–æ—Ä–µ–Ω–Ω—è –ø—Ä–æ—Å—Ç–æ—ó –≥—Ä–∏

```python
from abc import ABC, abstractmethod
from random import randint

class Item(ABC):
    def __init__(self, name:str, health = 500):
        self.name = name
        self.health = health
    
    @abstractmethod
    def attack(self):
        pass

class Sword(Item): # –¶–µ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–Ω—è
    def __init__(self, name, attack_power:int):
        super().__init__(name=name) #  –¶–µ –≤–∏–∫–ª–∏–∫ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∞ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–æ–≥–æ –∫–ª–∞—Å—É
        self.__attack_power = attack_power # –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç —è–∫–∏–π —Ä–µ–∞–ª—ñ–∑—É—î —ñ–Ω–∫–∞–ø—Å—É–ª—è—Ü—ñ—é
        self._sharp = 0
    
    def attack(self, another_item:Item): # –º–∏ –Ω–µ –º–æ–∂–µ–º–æ —Å—Ç–≤–æ—Ä–∏—Ç–∏ –∫–ª–∞—Å –±–µ–∑ —Ü—å–æ–≥–æ –º–µ—Ç–æ–¥—É
        current_attack = self.__attack_power + self._sharp + randint(0, 10)
        another_item.health -= current_attack
        return f"–ó–∞–≤–¥–∞—î–º–æ —É–¥–∞—Ä—É –º–µ—á–µ–º {self.name} —Ç–∞ –Ω–∞–Ω–æ—Å–∏–º–æ {current_attack} —à–∫–æ–¥–∏. –£ {another_item.name} –∑–∞–ª–∏—à–∞–ª–æ—Å—å –∑–¥–æ—Ä–æ–≤—è: {another_item.health}"
    
    @property
    def get_attack_power(self):
        return f"–ê—Ç–∞–∫–∞ –º–µ—á–∞ {self.name}: {self.__attack_power + self._sharp} –æ–¥–∏–Ω–∏—Ü—å"
    
    def sharpening(self):
        self._sharp += 1

class Axe(Item): # –¶–µ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–Ω—è
    def __init__(self, name, attack_power:int):
        super().__init__(name=name) #  –¶–µ –≤–∏–∫–ª–∏–∫ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ç–æ—Ä–∞ –Ω–∞—Å–ª—ñ–¥—É–≤–∞–Ω–æ–≥–æ –∫–ª–∞—Å—É
        self.__attack_power = attack_power # –ø—Ä–∏–≤–∞—Ç–Ω–∏–π –∞—Ç—Ä–∏–±—É—Ç —è–∫–∏–π —Ä–µ–∞–ª—ñ–∑—É—î —ñ–Ω–∫–∞–ø—Å—É–ª—è—Ü—ñ—é
        self._sharp = 0
    
    def attack(self, another_item:Item):  # –º–∏ –Ω–µ –º–æ–∂–µ–º–æ —Å—Ç–≤–æ—Ä–∏—Ç–∏ –∫–ª–∞—Å –±–µ–∑ —Ü—å–æ–≥–æ –º–µ—Ç–æ–¥—É
        current_attack = self.__attack_power + randint(0, 20)
        another_item.health -= current_attack
        return f"–ó–∞–≤–¥–∞—î–º–æ —É–¥–∞—Ä—É —Å–æ–∫–∏—Ä–æ—é {self.name} —Ç–∞ –Ω–∞–Ω–æ—Å–∏–º–æ {current_attack} —à–∫–æ–¥–∏. –£ {another_item.name} –∑–∞–ª–∏—à–∞–ª–æ—Å—å –∑–¥–æ—Ä–æ–≤—è: {another_item.health}"

    @property
    def get_attack_power(self):
        return f"–ê—Ç–∞–∫–∞ —Å–æ–∫–∏—Ä–∏ {self.name}: {self.__attack_power + self._sharp} –æ–¥–∏–Ω–∏—Ü—å"

S = Sword("–ï—Å–∫–∞–ª—ñ–±—É—Ä", 100)
A = Axe("–ö—Ä–∞—Ç–æ—Å", 95)

for i in range(10):
    print(f"–•—ñ–¥ {i}")

    S.sharpening()
    print(S.attack(A))
    if S.health <= 0:
        print(f"–ü–µ—Ä–µ–º–æ–≥–∞ –∑–∞ {A.name}")
        break

    print(A.attack(S))
    if A.health <= 0:
        print(f"–ü–µ—Ä–µ–º–æ–≥–∞ –∑–∞ {S.name}")
        break
```

# –î–æ–¥–∞–π—Ç–µ —Ç—Ä–µ—Ç—ñ–π —Ç–∏–ø –∑–±—Ä–æ—ó ‚Äî Bow (–ª—É–∫), —è–∫–∏–π:

- –º–∞—î –≤–ª–∞—Å–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä range_power (–¥–∞–ª—å–Ω—ñ—Å—Ç—å);
- –º–µ—Ç–æ–¥ attack() –∑ —Ñ–æ—Ä–º—É–ª–æ—é —à–∫–æ–¥–∏ attack_power + randint(5, 15) + range_power;
- –º–∞—î –º–µ—Ç–æ–¥ reload(), —è–∫–∏–π –∑–±—ñ–ª—å—à—É—î –¥–∞–ª—å–Ω—ñ—Å—Ç—å (range_power += 1); –î–æ–¥–∞–π—Ç–µ –≤–∏–ø–∞–¥–∫–æ–≤–∏–π –≤–∏–±—ñ—Ä –∑–±—Ä–æ—ó –∑ —Ç—Ä—å–æ—Ö –º–æ–∂–ª–∏–≤–∏—Ö (Sword, Axe, Bow); 
- –ó—Ä–æ–±—ñ—Ç—å –≥—Ä—É –ø–æ–∫—Ä–æ–∫–æ–≤–æ—é, –¥–µ –∫–æ—Ä–∏—Å—Ç—É–≤–∞—á –º–æ–∂–µ —Ä–æ–±–∏—Ç–∏ —Ö—ñ–¥ —Ç–∞ –≤–∏–±–∏—Ä–∞—Ç–∏ –¥—ñ—ó –∑ —Å–≤–æ—ó–º —Ç–∏–ø–æ–º –∑–±—Ä–æ—ó (–Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –∞—Ç–∞–∫—É–≤–∞—Ç–∏ –∞–±–æ –Ω–∞–∫–ª–∞—Å—Ç–∏ –ø—ñ–¥—Å–∏–ª–µ–Ω–Ω—è —â–æ–± –ø—ñ–¥–≤–∏—â–∏—Ç–∏ –µ—Ñ–µ–∫—Ç–∏–≤–Ω—ñ—Å—Ç—å –∑–±—Ä–æ—ó –Ω–∞ –Ω–∞—Å—Ç—É–ø–Ω–æ–º—É –∫—Ä–æ—Ü—ñ);

```python
from abc import ABC, abstractmethod
from random import randint, choice

class Item(ABC):
    def __init__(self, name: str, health=500):
        self.name = name
        self.health = health
        self._boost = 0  # —Ç–∏–º—á–∞—Å–æ–≤–µ –ø—ñ–¥—Å–∏–ª–µ–Ω–Ω—è

    @abstractmethod
    def attack(self, another_item):
        pass

    def apply_boost(self):
        self._boost += 5
        return f"{self.name} –æ—Ç—Ä–∏–º—É—î –ø—ñ–¥—Å–∏–ª–µ–Ω–Ω—è (+5 –¥–æ –Ω–∞—Å—Ç—É–ø–Ω–æ—ó –∞—Ç–∞–∫–∏)"


class Sword(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self._sharp = 0

    def attack(self, another_item):
        current_attack = (
            self.__attack_power + 
            self._sharp + 
            self._boost +
            randint(0, 10)
        )
        self._boost = 0
        another_item.health -= current_attack
        return f"‚öîÔ∏è {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏. –£ {another_item.name} HP: {another_item.health}"

    def sharpening(self):
        self._sharp += 1
        return f"{self.name} –∑–∞—Ç–æ—á–µ–Ω–æ. Sharpness: {self._sharp}"


class Axe(Item):
    def __init__(self, name, attack_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power

    def attack(self, another_item):
        current_attack = (
            self.__attack_power +
            self._boost +
            randint(0, 20)
        )
        self._boost = 0
        another_item.health -= current_attack
        return f"ü™ì {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏. –£ {another_item.name} HP: {another_item.health}"


class Bow(Item):
    def __init__(self, name, attack_power: int, range_power: int):
        super().__init__(name=name)
        self.__attack_power = attack_power
        self.range_power = range_power

    def attack(self, another_item):
        current_attack = (
            self.__attack_power +
            randint(5, 15) +
            self.range_power +
            self._boost
        )
        self._boost = 0
        another_item.health -= current_attack
        return f"üèπ {self.name} –∑–∞–≤–¥–∞—î {current_attack} —à–∫–æ–¥–∏. –£ {another_item.name} HP: {another_item.health}"

    def reload(self):
        self.range_power += 1
        return f"{self.name} –ø—ñ–¥–≤–∏—â—É—î –¥–∞–ª—å–Ω—ñ—Å—Ç—å. Range power: {self.range_power}"
```

```python
def random_weapon(player_name):
    weapon_class = choice([Sword, Axe, Bow])

    if weapon_class == Sword:
        return Sword(player_name, 100)
    elif weapon_class == Axe:
        return Axe(player_name, 95)
    else:
        return Bow(player_name, 85, 3)
```

```python
player = random_weapon("–ì—Ä–∞–≤–µ—Ü—å")
enemy = random_weapon("–ë–æ—Ç")

print(f"–í–∞—à–∞ –∑–±—Ä–æ—è: {type(player).__name__}")
print(f"–ó–±—Ä–æ—è –≤–æ—Ä–æ–≥–∞: {type(enemy).__name__}")
print("="*40)

turn = 1

while player.health > 0 and enemy.health > 0:
    print(f"\n–•—ñ–¥ {turn}")
    print(f"–í–∞—à HP: {player.health} | HP –≤–æ—Ä–æ–≥–∞: {enemy.health}")
    
    action = input("–û–±–µ—Ä—ñ—Ç—å –¥—ñ—é (1 - –∞—Ç–∞–∫–∞, 2 - –ø—ñ–¥—Å–∏–ª–µ–Ω–Ω—è, 3 - —Å–ø–µ—Ü. –¥—ñ—è): ")

    if action == "1":
        print(player.attack(enemy))
    elif action == "2":
        print(player.apply_boost())
    elif action == "3":
        if isinstance(player, Sword):
            print(player.sharpening())
        elif isinstance(player, Bow):
            print(player.reload())
        else:
            print("–£ —Ü—ñ—î—ó –∑–±—Ä–æ—ó –Ω–µ–º–∞—î —Å–ø–µ—Ü. –¥—ñ—ó")
    else:
        print("–ù–µ–≤—ñ—Ä–Ω–∏–π –≤–∏–±—ñ—Ä")

    if enemy.health <= 0:
        print("\nüéâ –í–∏ –ø–µ—Ä–µ–º–æ–≥–ª–∏!")
        break

    # –•—ñ–¥ –±–æ—Ç–∞
    print(enemy.attack(player))
    if player.health <= 0:
        print("\nüíÄ –í–∏ –ø—Ä–æ–≥—Ä–∞–ª–∏!")
        break

    turn += 1
```
