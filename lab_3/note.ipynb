# Перша програма на ООП

```python

class MyName:
    """Опис класу / Документація
    """
    total_names = 0 #Class Variable

    def __init__(self, name=None) -> None:
        """Ініціалізація класу
        """
        self.name = name if name is not None else self.anonymous_user().name #Class attributes / Instance variables
        MyName.total_names += 1 #modify class variable
        self.my_id = self.total_names

    @property
    def whoami(self) -> str: 
        """Class property
        return: повертаємо імя 
        """
        return f"My name is {self.name}"
    
    @property
    def my_email(self) -> str:
        """Class property
        return: повертаємо емейл
        """
        return self.create_email()
    
    def create_email(self) -> str:
        """Instance method
        """
        return f"{self.name}@itcollege.lviv.ua"

    @classmethod
    def anonymous_user(cls):
        """Classs method
        """
        return cls("Anonymous")
    
    @staticmethod
    def say_hello(message="Hello to everyone!") -> str:
        """Static method
        """
        return f"You say: {message}"


print("Розпочинаємо створювати обєкти!")

names = ("Bohdan", "Marta", None)
all_names = {name: MyName(name) for name in names}

for name, me in all_names.items():
    print(f"""{">*<"*20}
This is object: {me} 
This is object attribute: {me.name} / {me.my_id}
This is {type(MyName.whoami)}: {me.whoami} / {me.my_email}
This is {type(me.create_email)} call: {me.create_email()}
This is static {type(MyName.say_hello)} with defaults: {me.say_hello()} 
This is class variable {type(MyName.total_names)}: from class {MyName.total_names} / from object {me.total_names}
{"<*>"*20}""")

print(f"We are done. We create {me.total_names} names! ??? Why {MyName.total_names}?")
```

- імутуємо - гравці починають робити ходи картами, і карти будуть взаємодіяти між собою, атака буде зменшувати жувучіть поки карта не вилітає

```python
def check_if_player_has_cards(list_of_cards: list) -> bool:
    return True if len(list_of_cards) >= 1 else False


for move in range(3):
    if check_if_player_has_cards(player_1):
        card_on_board_1 = player_1.pop()
    else:
        print("У гравця 1 не залишилось карт")
    
    if check_if_player_has_cards(player_2):
        card_on_board_2 = player_2.pop()
    else:
        print("У гравця 2 не залишилось карт")

    while card_on_board_1['durability'] > 0 or card_on_board_2['durability'] > 0:
        card_on_board_1['durability'] -= card_on_board_2["attack_power"]
        card_on_board_2['durability'] -= card_on_board_1["attack_power"]

        print(f"""
        у гравця 1 залишилось витривалості: {card_on_board_1['durability']}
        у гравця 2 залишилось витривалості: {card_on_board_2['durability']}
        """)
    # перевіряємо яка з карт програла
    if card_on_board_1['durability'] <= 0 and check_if_player_has_cards(player_1):
        print("Карта гравця 1 Вибуває, гравець витягує нову карту")
        card_on_board_1 = player_1.pop()
    elif card_on_board_2['durability'] <= 0 and check_if_player_has_cards(player_2):
        print("Карта гравця 2 Вибуває, гравець витягує нову карту")
        card_on_board_2 = player_2.pop()
    else:
        print("У гравців закінчились карти")
```

```python
class CardGame:
    def __init__(self):
        # задиємо атрибути класу
        self.name = random.choice(names)
        self.attack_power = random.randint(4, 8)
        self.durability = random.randint(30, 60)
        self.magic_types = random.choice(magic_types)
        self.description = "Це опис за замовчуванням"

    @property
    def basic_attack(self):
        crit_chance = random.random()
        if crit_chance > 0.8:
            return self.attack_power * 1.5
        return self.attack_power
    
    def attack(self, target: CardGame) -> str:
        # приватна змінна для збереження значення шкоди
        _damage = self.basic_attack
        target.durability -= _damage
        return f">-- {self.name} наніс {_damage} шкоди {target.name}"
    
    def heal_self(self) -> str:
        amount = random.randint(1, 3)
        self.durability += amount
        return f">++ {self.name} відновив {amount} витривалості"
```

```python
a = CardGame()
b = CardGame()

while a.durability > 0 and b.durability > 0:
    print(f"Карта 1: {a.name} з атакою {a.attack_power} і витривалістю {a.durability}")
    print(f"Карта 2: {b.name} з атакою {b.attack_power} і витривалістю {b.durability}")
    if random.random() > 0.5:
        print(a.heal_self())
    else:
        print(a.attack(b))
    
    if random.random() > 0.5:
        # обидва записи є коректними
        print(CardGame.heal_self(b))
        print(b.heal_self())
    else:
        print(b.attack(a))
```

```python
class CardGame:
    def __init__(self):
        # задиємо атрибути класу
        self.name = random.choice(names)
        print(f"Створюємо нову карту {self.name}...")
        self.attack_power = CardGame.generate_attack_power()
        self.durability = random.randint(30, 60)
        self.magic_types = random.choice(magic_types)
        self.description = CardGame.generate_card_type()

    @staticmethod
    def generate_attack_power():
        # генеруємо базову атаку з можливими бонусами
        _a = random.randint(5, 10)
        if random.random() > 0.8: # з шансом 20% буде буст атаки
            print("Застосували буст атаки!")
            _a += random.randint(1, 3)
        if random.random() > 0.9: # з шансом 10% буде додатковий удар
            print("Випав шанс на додатковий удар!")
            _a *= 1.2
        if random.random() > 0.95: # з шансом 5% буде критичний удар
            print("Випав шанс на критичний удар!")
            _a *= 2
        return _a
    
    @staticmethod
    def generate_card_type():
        _a = 0
        color = random.choice(['Червоний', 'Синій', 'Зелений', 'Жовтий'])
        match color:
            case 'Червоний':
                magic_types = 'Вогняна'
                _a += 2 + random.randint(1, 5) 
            case 'Синій':
                magic_types = 'Водяна'
                _a += 1 + random.randint(1, 5)
            case 'Зелений':
                magic_types = 'Вітеряна'
                _a += 3 + random.randint(1, 5)
            case 'Жовтий':
                magic_types = 'Блискавкова'
                _a += 4 + random.randint(1, 5)
        return f"Наповнена {color} блиском: Це потужна {magic_types} магія що додає {_a} шкоди"
        
    def __str__(self):
        return f"{self.name} | Атака: {self.attack_power} | Витривалість: {self.durability} | Магія: {self.magic_types} | Опис: {self.description}"
    
    def __len__(self):
        return self.durability
```

- класичиний приклад для альтернитивних конструкторів - Персона

```python
class Persona:
    def __init__(self, name: str, surname: str):
        # задиємо атрибути класу
        self.name = name
        self.surname = surname
    
    @classmethod
    def from_full_name(cls, full_name: str):
        name, surname = full_name.split(" ")
        return cls(name, surname)

    @classmethod
    def from_dict(cls, data: dict):
        return cls(data['name'], data['surname'])
    
    @classmethod
    def from_list(cls, data: list):
        return cls(data[0], data[1])
    
p1 = Persona("Іван", "Іванов")
p2 = Persona.from_full_name("Іван Іванов")
p3 = Persona.from_dict({"name": "Іван", "surname": "Іванов"})

print(p1.name, p1.surname, p2.name, p2.surname, p3.name, p3.surname)
```

```python
class InternalVariables:
    def simple_method(self):
        # локальні змінні методу
        local_var = "Я локальна змінна методу simple_method"
        print(local_var)
        # можна передати цю змінну через динамічний атрибут класу
        self.local_var = local_var
        # краще пеердавати такі змінні як приватні атрибути класу
        self._private_var = "Я приватна змінна класу"
        # захищені атрибути класу починаються з двох підкреслення
        self.__protected_var = "Я захищена змінна класу"
    
    def another_method(self):
        print("Вивід з another_method:", self.local_var)  # NameError: name 'local_var' is not defined
        a = self._private_method()
        print(f"Вивід після приватнрго меьтоду: {a}")
    
    @property
    def get_private_var(self):
        return self._private_var

    def _private_method(self):
        "Такі приватні методи бажано не викликати зовні класу"
        print("Це приватний метод класу")
        return 42

obj = InternalVariables()
obj.simple_method()
obj.another_method()

print("Доступ до приватної змінної:", obj.get_private_var, obj._private_var)
obj._private_var = "Змінили приватну змінну зовні класу"
print("Доступ до приватної змінної після зміни:", obj.get_private_var)
```
